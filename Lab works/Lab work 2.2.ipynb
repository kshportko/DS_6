```python
#"This function performs cleaning of the text and 
# it makes a list of words from cleaned text assuming that each word is 3 or more symbols long"
def text_clean(text):
    text=text.lower()
    punct=['.',',','—','!','?']
    for symbol in punct:
        if symbol in text:
            text=text.replace(symbol,"")
    list=[]
    list=text.split()       
    fine_list=[]
    for i in range(len(list)-1):
        if len(list[i])>=3:
            fine_list.append(list[i])
    return(fine_list)
```

```python
if __name__=='__main__':
    
    print('Number of words in the text: ', len(text_clean(text))) 
    print('Number of unique words in the text: ', len(set(text_clean(text))))
```

```python
#"This function returns a dictionary with words as keys and their frequencies as values"
def magic_dict(some_list):
    magic_dict={}
    for w in some_list:
        magic_dict[w]=magic_dict.get(w,0)+1
    return(magic_dict)
    
    
if __name__=='__main__':
    print(magic_dict(text_clean(text)))
    
```

```python
#"This function returs a dictionary with top N of words from the given dictionary"
def top_words(dict_words,n):
    res={}
    res1={}
    lis=sorted(set(dict_words.values()), reverse=True)
    if n<= len(lis)-1:
        lis=lis[0:n]
        for i, j in dict_words.items():
            if j in lis:
                res[i]=j
        sorted_keys=sorted(res, key=res.get, reverse=True)
        for w in sorted_keys:
            res1[w] = res[w]
        return(res1)
    else:
        res=None

if __name__=='__main__':
    n=int(input('Enter the number for the top list: '))
    for i, j in top_words(magic_dict(text_clean(text)),n).items():
        print(i, j)
    
```

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,99) for i in range(1000)]
```

```python
def min_max(list):
    list.sort()
    return('The minimum and maximum values are: {} and {}' .format(list[0], list[999]))


min_max(list_with_random_numbers)
```

```python
def average(list):
    res=0
    for i in range(len(list)):
        res+=list[i]
    res/=len(list)
    return(res)

average(list_with_random_numbers)       
```

```python
def mediana(list):
    res=0
    list1=[]
    list=set(list)
    for i in list:
        list1.append(i)
    list1.sort()
    
    if len(list1)%2==0:
        res=int((list1[int(len(list1)/2-1)]+list1[int(len(list1)/2)])/2)
    else:
        res=list1[int(len(list1)//2+1)]
        
    return(res)


mediana(list_with_random_numbers)  
```

```python
def quadr_dev(list):
    import math
    res=0
    av=0
    for i in range(len(list)):
        av+=list[i]
    av/=len(list)
    for i in range(len(list)):
        res+=(list[i]-av)**2
    res=math.sqrt(res/(len(list)-1))
    return(res)

quadr_dev(list_with_random_numbers)      
```
