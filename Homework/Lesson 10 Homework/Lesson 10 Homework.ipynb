```python
HEADERS = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:49.0) Gecko/20100101 Firefox/49.0',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.5',
            'Accept-Encoding': 'gzip, deflate',
            'DNT': '1',
            'Connection': 'keep-alive',
            'Upgrade-Insecure-Requests': '1',
          }
```

```python
import requests
from bs4 import BeautifulSoup

SITE = 'https://www.ukr.net/news/main.html'

r = requests.get(SITE, headers=HEADERS)

if r.status_code == 200:
    soup = BeautifulSoup(r.text, 'lxml')
    for el in soup.find_all('a', class_='n-m_li_a ps-h'):
        print('\n"')
        print(el. text, el.get('href'))
        SITE2 = 'https:'+el.get('href')
        r2 = requests.get(SITE2, headers=HEADERS)
        if r2.status_code == 200:
            soup2 = BeautifulSoup(r2.text, 'lxml')
            
            for el2 in soup.find_all('section', class_='im'):
                time=el2.find('time', class_='im-tm').text
                title=el2.find('a', class_='im-tl_a').text
                print(time, title)
```
